# Printing

```julia
# Another way to format strings is the printf macro.
@printf "%d is less than %f\n" 4.5 5.3 # 5 is less than 5.300000

# Printing is easy
println("I'm Julia. Nice to meet you!")
```

# Loops

```julia
# Julia works much the same as Python, syntax-wise
for i in 1:3
    println(i) #indent not needed; Use $i to print variable when in quotes
    println("i = $i")
    @printf "i = %d\n" i #print macro w/c-like formatting
end
```

```julia
words = ["Luke", "Leia"] #define a word array
for word in words
    println("Hello, $word")
end
```

# Plotting

```julia
# To use a package (which we have installed using Pkg.Add("Name")),
using PyPlot
```

```julia
# Set number of points 
n = 50

# Set random number seed
srand(1)

# Draw n random numbers
x = rand(n)
y = rand(n)

# Calulate a random area
area = pi .* (15 .* rand(n)).^2 # 0 to 15 point radii

# Plot
scatter(x, y, s=area, alpha=0.5) #transparency, circle sizes = area
xlabel("x Value")
ylabel("y Value")
```

```julia
ts_length = 100
# Draw numbers from normal distribution
epsilon_values = randn(ts_length)
plot(epsilon_values, "k-") #plot black line (k-)
```
